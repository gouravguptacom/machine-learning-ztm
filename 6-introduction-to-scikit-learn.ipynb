## 0. An end-to-end Sciket-Learn workflow

```python
# Standart Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Get the data ready
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# Create x (feature matrix)
x = heart_disease.drop("target", axis=1)

# Create y (labels)
y = heart_disease["target"]
```

```python
# 2. Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)

# We'll keep the default hyperparameters
clf.get_params()
```

```python
# 3. Fit the model to the training data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
clf.fit(x_train, y_train) #;
```

```python
x_train
```

```python
import numpy as np

# make a prediction
# y_label = clf.predict(np.array([0, 2, 3, 4]))
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improve a model
# Try different amount of n_estimators
np.random.seed(42)

for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accuracy on test set: {clf.score(x_test, y_test) * 100:.2f}%")
    print("")
```

```python
# 6. Save a model and load it

import pickle

pickle.dump(clf, open("random_forst_model_1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random_forst_model_1.pkl", "rb"))
loaded_model.score(x_test, y_test)
```

### 1.1 Make sure it's all numerical

```python
car_sales = pd.read_csv("car-sales-extended.csv")
car_sales.head()
```

```python
len(car_sales)
```

```python
car_sales.dtypes
```

```python
# Split into X/y
X = car_sales.drop("Price", axis=1)
y = car_sales["Price"]

# Split into training and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
# Build machine learning model
from sklearn.ensemble import RandomForestRegressor # this can predict a number, unlike classifier which predict yes or no

model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# Before we run machine learning models, we need to turn these into numbers
```

```python
# Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]

# Car Door's are also categorical because we have x amount of cars which has 4 doors

one_hot = OneHotEncoder()
transformer = ColumnTransformer([( 
        "one_hot", 
        one_hot, 
        categorical_features
    )], 
    remainder="passthrough"
)

transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
pd.DataFrame(transformed_X)
```
